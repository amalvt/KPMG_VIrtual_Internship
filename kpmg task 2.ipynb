# identifying the trends and pattern from transactions table 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
transactions=pd.read_csv('transactions.csv')
```

```python
transactions
```

checking the trend with buying online

```python
status_counts=transactions['online_order'].value_counts()

# #plotting the graph
status_counts.plot.pie(figsize=(4, 3), autopct='%1.1f%%')
plt.ylabel('')
plt.title('Online Order Status')
plt.show()
```

# Identifing the trends and pattern from Customerdemographic table 

```python
cusdemo=pd.read_csv('cusdemo.csv')
```

```python
cusdemo
```

```python
counts=cusdemo['gender'].value_counts()

# #plotting the graph
counts.plot.pie(figsize=(4, 3), autopct='%1.1f%%')
plt.show()
```

Identifying the buying pattern for different age groups 

```python
# Defining the age groups and corresponding labels
age_groups = [0, 20, 30, 40, 50, 60, float('inf')]
age_labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60+']

# Categorizing the 'age' column into age groups
cusdemo['age_group'] = pd.cut(cusdemo['age'], bins=age_groups, labels=age_labels, right=False)

# Group by age group and calculate the sum of purchases
age_group_purchases = cusdemo.groupby('age_group')['past_3_years_bike_related_purchases'].sum()

# Plotting the bar chart
age_group_purchases.plot.bar(figsize=(8, 5))
plt.xlabel('Age Group')
plt.ylabel('Total Purchases')
plt.title('Distribution of Purchases by Age Group')
plt.show()
```

from this it is clear that people aged 40-49 have contributed more in buying bicycles 

..................................................................................................................................     

identifying the pattern with job industry catogery

```python
#grouping by gender and calculating the sum of purchases
job_catog=cusdemo.groupby('job_industry_category')['past_3_years_bike_related_purchases'].sum()

#plotting the bar chart 
job_catog.plot.bar(figsize=(8,5))
plt.xlabel('job_industry_catogory')
plt.ylabel('Total purchase')
plt.title('Distribution of Purchases by job_industry_catogory')
plt.show()
```

it is clear that people working in financial services ,health and manufacturing have contributed more 

.................................................................................................................................

identifying the pattern associated with age groups and job industry catogery

```python
age_group_counts = cusdemo.groupby('job_industry_category')['age_group'].value_counts().unstack().fillna(0)
age_group_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Job Industry Category')
plt.ylabel('Count')
plt.title('Age Group Distribution by Job Industry Category')
plt.legend(title='Age Group')
plt.show()
```

identifying the pattern assiciated with having and not having a car 

```python
#grouping by owns_Car and calculating the sum of purchases
purchases=cusdemo.groupby('owns_car')['past_3_years_bike_related_purchases'].sum()

#plotting the graph
purchases.plot.pie(figsize=(4, 3), autopct='%1.1f%%')
plt.show()
```

analysing the pattern for people in different welath segment 

```python
#grouping by wealth_segment and calculating the sum of purchases
segment_purchases=cusdemo.groupby('wealth_segment')['past_3_years_bike_related_purchases'].sum()

#plotting the graph
segment_purchases.plot.pie(figsize=(4, 3), autopct='%1.1f%%')
plt.show()
```

```python
segment_counts = cusdemo.groupby('wealth_segment')['age_group'].value_counts().unstack().fillna(0)
segment_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('welath_segment')
plt.ylabel('Count')
plt.title('Age Group Distribution by welath_segment')
plt.legend(title='Age Group')
plt.show()
```
