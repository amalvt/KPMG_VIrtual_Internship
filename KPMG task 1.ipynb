# Transactions table  quality assessment

```python
#loading the Transactions table
transactions=pd.read_excel('KPMG.xlsx',sheet_name='Transactions')
transactions
```

```python
transactions.info()
```

```python
# changing the product_first_sold_date format 
transactions['product_first_sold_date'] = pd.to_datetime(transactions['product_first_sold_date'], format='%Y-%m-%d').dt.date
transactions
```

```python
# Checking for null values 
transactions.isna().sum()
```

null values are present in online_order,brand,product_line,product_class ,product_size,standard_cost,product_first_sold_date columns and can be renmoved or replaced in the further analyis

```python
#checking for duplicate values 

duplicates = transactions.duplicated()
transactions[duplicates].sum()
```

No duplicate values are present

```python
transactions.columns
```

Exploring Through columns

```python
for column in transactions.columns:
    value_counts = transactions[column].value_counts()
    print(f"Value counts for column '{column}':\n{value_counts}\n")
    print('........................................................')
```

Creating a new column profit 

```python
transactions['profit'] = transactions['list_price'] - transactions['standard_cost']
transactions
```

Saving it to a csv file

```python
transactions.to_csv('transactions.csv',index=False)
```

# Newcustomerlist table quality assessment

```python
#loading the Transactions table
newcus=pd.read_excel('KPMG.xlsx',sheet_name='NewCustomerList')
newcus
```

```python
newcus.info()
```

```python
newcus.columns
```

columns 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',  'Unnamed: 19', 'Unnamed: 20' are irrelevnt for the analysis hence the can be dropped 

```python
# dropping  columns 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',  'Unnamed: 19', 'Unnamed: 20'
columns_to_drop = ['Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18','Unnamed: 19', 'Unnamed: 20']
newcus = newcus.drop(columns=columns_to_drop)
```

```python
newcus
```

```python
# Checking for null values 
newcus.isna().sum()
```

Null values are present in columns - last_name,DOB,job_title,job_industry_catogery and they can be dropped or replaced in the further steps of analysis

```python
#checking for duplicate values
newcus.duplicated().sum()
```

no duplicate values are present

Exploring through columns

```python
for column in newcus.columns:
    value_counts = newcus[column].value_counts()
    print(f"Value counts for column '{column}':\n{value_counts}\n")
    print('........................................................')
```

Replacing U in gender with unspecified 

```python
#Replacing U in gender with unspecified 
newcus['gender']=newcus['gender'].replace({'U':'Unspecified'})
```

Creating a new column for age 


```python
newcus['DOB'] = pd.to_datetime(newcus['DOB'])  # Convert 'dob' column to datetime type
newcus['age'] = (datetime.now() - newcus['DOB']).astype('<m8[Y]')  # Calculate age based on current date
```

```python
newcus['age'].value_counts()
```

```python
newcus
```

saving to csv file

```python
newcus.to_csv('newcus.csv',index=False)
```

# Customerdemographic table quality assessment

```python
cusdemo=pd.read_excel('KPMG.xlsx',sheet_name='CustomerDemographic')
cusdemo
```

```python
cusdemo.info()
```

```python
#checking for null values
cusdemo.isna().sum()
```

null values are present in last_name,DOB,job_title,job_industry_catogery and tenure columns and can be removed or replaced in the further analysis 

```python
#checking for duplicate values
cusdemo.duplicated().sum()
```

no duplicate values are present

Exploring through columns

```python
for column in cusdemo.columns:
    value_counts = cusdemo[column].value_counts()
    print(f"Value counts for column '{column}':\n{value_counts}\n")
    print('........................................................')
```

Corrections are done to values F,M and U of Gender column

```python
#replacing the values F and Femal with Female,M with male and U with unspecified 
cusdemo['gender']=cusdemo['gender'].replace({'F':'Female','Femal':'Female','M':'Male','U':'Unspecified'})
```

default column can be removed as it contain irrelevent content for the analysis 

```python
#dropping the column default as it contains irrelevent content with symbals 
del cusdemo['default']
```

Creating new column age

```python
# creating new column age
cusdemo['DOB'] = pd.to_datetime(cusdemo['DOB'])  # Convert 'dob' column to datetime type
cusdemo['age'] = (datetime.now() - cusdemo['DOB']).astype('<m8[Y]')  # Calculate age based on current date
```

```python

cusdemo['age'].value_counts()
```

```python
# dropping the age 179 as it is not accurate data 
cusdemo = cusdemo.drop(cusdemo[cusdemo['age'] == 179].index)
```

```python
cusdemo
```

saving to csv file 

```python
cusdemo.to_csv('cusdemo.csv',index=False)
```
